```python
import pandas as pd
import quandl
import numpy as np
import math 
from sklearn import preprocessing, cross_validation, svm 
from sklearn.linear_model import LinearRegression
```

```python
df = quandl.get('WIKI/GOOGL')
df
```

```python
df.tail()
```

```python
df.info()
```

```python
## creating a new df with subsets. 
## A little bit about Adj, since the split ratio is 1, this just means that after
## stock splits 

## The margin of High and Low tells us a bit about the volatiltiy of the day

df = df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]
df.head()
```

```python
# Let's define some features... 

# High - Low percent

df['HL_PCT'] = (df['Adj. High'] - df['Adj. Low'])/df['Adj. Low'] *100.0

## Percent Change --  close - open
df['PCT_change'] = (df['Adj. Close'] - df['Adj. Open'])/df['Adj. Open'] *100.0
```

```python
df.head()
```

```python
df = df[['Adj. Close','HL_PCT','PCT_change','Adj. Volume']]
```

```python
df.head()
```

```python
forecast_col = 'Adj. Close'
#df.fillna(-99999, inplace = True)

forecast_out = int(math.ceil(0.01*len(df)))
forecast_out
```

```python
df['label'] = df[forecast_col].shift(-forecast_out)
```

```python
df.head()
```

```python
df.tail()
```

```python
## As we can see the last rows have no values.. since there are no inputs for it

df.tail(50)
```

```python
df.dropna(inplace= True)
```

```python
# All rows up to NA have now been dropped. 

df.tail(10)
```

```python
X = np.array(df.drop(['label']))
y = np.array(df['label'])
```

```python
## Creating train X and Train Y

X = np.array(df.drop(['label'],1))
y = np.array(df['label'])
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size =0.2)
```

```python
X_train
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
## Accuracy of the data... R^2

clf = LinearRegression()
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
accuracy
```

```python
## SVM 

clf = svm.SVR()
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
accuracy
```

```python
## svm.LinearRegression

clf = svm.LinearSVR()
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
accuracy
```
